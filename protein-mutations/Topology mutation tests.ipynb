```python
pdbfile = app.PDBFile('p-xylene/receptor.pdb')
topology = pdbfile.topology

# Extract reference residue.
reference_residue_index = 110 # index of residue to mutate (Python numbering)
residues = [ residue for residue in topology.residues() ]
reference_residue = residues[reference_residue_index]

print "prev> " + residues[109].name 
print "====> " + residues[110].name + " *"
print "next> " + residues[111].name

# Retrieve templates
templates = forcefield._templates

# Find external bonds from reference residue.
prev_residue_link_atom = None
next_residue_link_atom = None
reference_residue_atoms = [ atom for atom in reference_residue.atoms() ] 
for (iatom,jatom) in topology._bonds:
    if (iatom in reference_residue_atoms) or (jatom in reference_residue_atoms):
        if (iatom.residue.index < reference_residue_index):
            prev_residue_link_atom = iatom
        if (jatom.residue.index < reference_residue_index):
            prev_residue_link_atom = jatom        
        if (iatom.residue.index > reference_residue_index):
            next_residue_link_atom = iatom
        if (jatom.residue.index > reference_residue_index):
            next_residue_link_atom = jatom
link_atoms = list()
if prev_residue_link_atom: link_atoms.append(prev_residue_link_atom)
if next_residue_link_atom: link_atoms.append(next_residue_link_atom)

# Add additional residues to specified residue index.
additional_resnames = ['ALA', 'ILE', 'PHE'] # list of residues to add
for resname in additional_resnames:
    # Create residue
    residue = app.topology.Residue(resname, reference_residue.index, reference_residue.chain)
    # Add atoms.
    template = templates[resname]
    new_atoms = list()
    for template_atom in template.atoms:
        atom = topology.addAtom(template_atom.name, template_atom.element, residue)
        new_atoms.append(atom)
    # Add bonds.
    for (i,j) in template.bonds:
        topology.addBond(new_atoms[i], new_atoms[j])
    # Connect up to residues on either side, assuming external bonds should be preserved.
    for (link_index, local_index) in enumerate(template.externalBonds):
        topology.addBond(link_atoms[link_index], new_atoms[local_index])
    
    # Add residue.
    reference_residue.chain._residues.append(residue)    
```
